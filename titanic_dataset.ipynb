```python
# Titanic Survival Analysis - Optimized Code

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and inspect data
def load_data():
    url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
    df = pd.read_csv(url)
    print("Dataset loaded successfully!")
    return df

titanic = load_data()

# Initial data inspection
def data_overview(df):
    print("\n=== DATA OVERVIEW ===")
    print(f"Shape: {df.shape}")
    print("\nFirst 5 rows:")
    display(df.head())
    
    print("\n=== DATA TYPES ===")
    print(df.info())
    
    print("\n=== STATISTICAL SUMMARY ===")
    display(df.describe(include='all'))
    
    print("\n=== MISSING VALUES ===")
    display(df.isnull().sum().to_frame('Missing Values'))

data_overview(titanic)

# Data cleaning
def clean_data(df):
    # Fill age with median (more robust than mean)
    df['Age'].fillna(df['Age'].median(), inplace=True)
    
    # Drop cabin (too many missing values) and other less useful columns
    df.drop(columns=['Cabin', 'PassengerId', 'Ticket'], inplace=True)
    
    # Fill embarked with mode
    df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
    
    # Create family size feature
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
    
    # Create age groups
    bins = [0, 12, 18, 30, 50, 100]
    labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
    df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
    
    return df

titanic = clean_data(titanic)

# Survival analysis
def analyze_survival(df):
    print("\n=== SURVIVAL ANALYSIS ===")
    
    # Overall survival rate
    overall_survival = df['Survived'].mean() * 100
    print(f"\nOverall survival rate: {overall_survival:.2f}%")
    
    # By gender
    gender_survival = df.groupby('Sex')['Survived'].mean() * 100
    print("\nSurvival by gender:")
    display(gender_survival.to_frame('Survival Rate (%)'))
    
    # By passenger class
    class_survival = df.groupby('Pclass')['Survived'].mean() * 100
    print("\nSurvival by passenger class:")
    display(class_survival.to_frame('Survival Rate (%)'))
    
    # By age group
    age_survival = df.groupby('AgeGroup')['Survived'].mean() * 100
    print("\nSurvival by age group:")
    display(age_survival.to_frame('Survival Rate (%)'))
    
    # Family size impact
    family_survival = df.groupby('FamilySize')['Survived'].mean() * 100
    print("\nSurvival by family size:")
    display(family_survival.to_frame('Survival Rate (%)'))
    
    # Youngest and oldest survivors
    survivors = df[df['Survived'] == 1]
    youngest = survivors.sort_values('Age').iloc[0]
    oldest = survivors.sort_values('Age', ascending=False).iloc[0]
    
    print(f"\nYoungest survivor: {youngest['Name']} ({youngest['Age']:.2f} years)")
    print(f"Oldest survivor: {oldest['Name']} ({oldest['Age']:.2f} years)")

analyze_survival(titanic)

# Visualization
def create_visualizations(df):
    plt.figure(figsize=(15, 10))
    
    plt.subplot(2, 2, 1)
    sns.barplot(x='Pclass', y='Survived', data=df)
    plt.title('Survival Rate by Passenger Class')
    
    plt.subplot(2, 2, 2)
    sns.barplot(x='Sex', y='Survived', data=df)
    plt.title('Survival Rate by Gender')
    
    plt.subplot(2, 2, 3)
    sns.barplot(x='AgeGroup', y='Survived', data=df, order=labels)
    plt.title('Survival Rate by Age Group')
    plt.xticks(rotation=45)
    
    plt.subplot(2, 2, 4)
    sns.barplot(x='Embarked', y='Survived', data=df)
    plt.title('Survival Rate by Embarkation Port')
    
    plt.tight_layout()
    plt.savefig('titanic_survival_analysis.png')
    plt.show()

create_visualizations(titanic)
```
